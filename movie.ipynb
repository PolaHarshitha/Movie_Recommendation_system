```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")

print("Path to dataset files:", path)
```

```python
import os
files=os.listdir(path)
print(files)
```

```python
import pandas as pd
credits = pd.read_csv(os.path.join(path,"tmdb_5000_credits.csv"))
movies = pd.read_csv(os.path.join(path,"tmdb_5000_movies.csv"))
```

```python
credits.head()
```

```python
movies.head()
```

```python
movies=movies.merge(credits,left_on="title",right_on="title")
```

```python
movies.head(1)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(1)
```

```python
''' Converts a stringified list of dictionaries (from CSV) into a Python list of only the 'name' values for easier 
data processing. '''
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head(1)
```

```python
movies['cast']=movies['cast'].apply(convert)
movies.head(1)
```

```python
movies['crew']=movies['crew'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(1)
```

```python
movies['tags']=movies['tags'].apply(lambda x: " ".join(x))
movies.head(1)
```

```python
movies = movies[['movie_id','title','overview','tags']]
movies.head(1)
```

```python
'''  Converts text into a numerical representation (TF-IDF values).
Ignores common English words like “the”, “and”, “is” (stop_words='english').
fit_transform(movies['tags'])
fit learns the vocabulary from your tags column.
transform converts each movie’s tags into a vector of TF-IDF scores.'''

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['tags'])
```

```python
''' This computes cosine similarity between every pair of movies based on their TF-IDF vectors.
The result is a similarity matrix where higher values mean movies have more similar content (used for recommendations).'''
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movies[movies['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10 similar movies
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]
```

```python
print(get_recommendations('The Dark Knight Rises'))
```
